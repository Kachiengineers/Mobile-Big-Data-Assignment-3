# Part 1: Violent and Non-Violent Extremism

**1. Violent vs Non-Violent Extremism**

In this part of the assignment, we extend our earlier extremism detection
approach by distinguishing between violent extremism (VE) and non-violent
extremism (NVE). Rather than treating extremism as a single category, this
allows us to capture meaningful differences in how harmful content manifests.
Knight, Woodward and Lancaster (2017) note that both groups share a common
ideological foundation, and that many factors commonly associated with
extremism, such as identity seeking, a sense of purpose, and outgroup
hostility, are present in both VEs and NVEs at comparable rates and therefore
do not distinguish between the two. In fact, all 40 case studies in their
sample, both violent and nonviolent, showed evidence of perceiving an external
threat and seeking like-minded others, confirming that ideology and grievance
alone are insufficient predictors of violence.

The main distinction lies in the presence of violence and the intent to cause
physical harm. Knight et al. (2017) found a specific set of statistically
significant variables that separate VEs from NVEs, including personal
responsibility to act, exposure to extreme violence, experiences of bullying,
low self-esteem, underachievement, and participation in extremist-related
training and travel. Notably, significantly more VEs felt a personal
responsibility to act (87.5% vs 18.8%), which the authors describe as a key
factor in determining violent action. NVE, in contrast, involves promoting
extreme ideologies or supporting extremist causes through non-physical means,
such as disseminating propaganda or fundraising, without direct involvement in
or facilitation of violence. While NVE may appear less severe, the authors
acknowledge that some individuals categorized as NVE may be pre-violent rather
than genuinely nonviolent, meaning they had not yet acted violently but could
be on a pathway toward it. This motivates the need to treat VE and NVE as
related but distinct categories.

**2. Definitions of Extremism and Radicalization**

Knight et al. (2017) devote considerable attention in their method section to
the problem of definitional inconsistency in the literature, noting that many
studies fail to clearly define extremism or conflate extremist attitudes with
violent action. For their study, they define extremists pragmatically as
individuals known to hold attitudes and beliefs that do not fit with mainstream
opinion on political or ideological issues, who had been convicted of a
criminal offense where there was evidence of extremist motivation. This
conviction-grounded, behavior-based definition was chosen to support
counter-terrorism practitioners rather than to resolve theoretical debates.

On radicalization, the authors draw on the UK Home Office definition, which
frames it as the process by which people come to support terrorism and
extremism and, in some cases, to then participate in terrorist activity.
However, they are careful to note that radicalization should not be assumed to
inevitably lead to violence, stating that it is only one of various pathways
into violent extremism. Crucially, their study does not examine how people
become radicalized. Instead, the focus is on how, when and why individuals
choose different extremist-related actions, whether violent or nonviolent.
Radicalization is therefore treated as contextual background rather than the
primary object of study, with the analytical emphasis placed on observable
behavioral and psychological variables that distinguish outcomes.

**3. Definition of VE vs NVE Used in Our Approach**

Following Knight et al. (2017), violent extremism (VE) is defined as any act,
or any clearly intended act, constituting murder, attempted murder, assault,
or serious physical harm to persons or property in pursuit of an ideological
goal. This definition also extends to individuals who knowingly conduct
nonviolent behaviors that facilitate violence by others, such as financing
attacks or providing operational support, who are still categorized as VEs. In
the study, VE offenses included terrorist attacks or foiled plots intended to
kill others.

Non-violent extremism (NVE) refers to the promotion, dissemination or support
of an extreme ideology through non-physical means, without direct involvement
in or facilitation of violence. In the study, NVE offenses included possession
of or dissemination of extremist literature, sending racist hate mail, and
fundraising for extremist organizations. Importantly, both groups share
ideological content and grievances, meaning classification must focus on
behavior and intent toward violence rather than ideology alone.

These definitions were incorporated directly into our prompt to ensure that
the classification process remains consistent with the empirical distinctions
established by Knight et al. (2017), as follows:

> **Prompt:** You are a content classifier trained in extremism research.
> Classify the following text as either Violent Extremism (VE) or Non-Violent
> Extremism (NVE) using the definitions below.
>
> **VE:** Content or behavior involving any act, or clear intent to commit an
> act, constituting murder, attempted murder, assault, or serious physical harm
> to persons or property in pursuit of an ideological goal. This also includes
> content that knowingly facilitates such violence by others, such as financing
> attacks, supplying weapons, or operational planning for violence.
>
> **NVE:** Content or behavior that promotes, disseminates or supports an
> extreme ideology without involving or facilitating direct physical harm. This
> includes producing or distributing extremist literature, sending hate
> material, fundraising for extremist causes, or recruiting others to an
> extreme ideology, where there is no direct link to planned or actual physical
> violence.
>
> **Important:** Both VE and NVE share ideological content, grievances and
> outgroup hostility. These features alone do not distinguish the two. Classify
> based on whether the content involves, plans or enables physical harm (VE),
> or whether it is confined to ideological expression or material support
> without a direct violence component (NVE). Content that funds or logistically
> supports a violent act should be classified as VE even if the author did not
> personally commit violence. Output VE or NVE, followed by one sentence
> identifying the specific indicator that drove the classification.



**4. Prompt Design Choices**

We designed our prompt to produce structured and interpretable outputs across
multiple dimensions. Instead of allowing free-form responses, we required the
model to return results in a fixed JSON format. This decision simplifies
downstream processing and ensures consistency when handling a large number of
examples.

The three VE subtypes used in our classification: Ideological, Political and
Religious, are grounded in the broader extremism literature. Ismail, Jamir
Singh and Mujani (2025) identify belief and behaviour as the two dominant
dimensions across which extremism manifests, with rigid ideological conviction,
biased interpretation of doctrine, and context-specific behavioural expression
forming the core analytical categories. This maps directly onto the ideological,
religious and political subtype distinctions used in our prompt. Berhoum et al.
(2023) similarly demonstrate that extremism in social media content can be
meaningfully classified across subtypes using machine learning approaches,
validating the feasibility of automated subtype classification from text alone.
While Knight et al. (2017) do not define these subtypes directly, they are a
natural extension of their framework for practitioners needing finer-grained
risk differentiation beyond the binary VE/NVE distinction.

Another important design choice was the inclusion of a KEY_SPAN field. This
requires the model to identify specific words or phrases that influenced its
decision, improving transparency and making it easier to verify whether the
classification is grounded in the input text. We also constrained the set of
possible target groups to those specified in the assignment, which helps reduce
ambiguity and ensures that outputs remain consistent across samples. Overall,
the prompt was designed to balance clarity for the model with usability for
analysis.

## Step 6 - Classification Function

Each post is classified across five dimensions in a single API call, returning a structured JSON object.

**Dimensions classified:**

| Field | Description |
|---|---|
| `VE_NVE` | Violent or Non-Violent Extremism |
| `VE_SUBTYPE` | Ideological / Political / Religious (VE only) |
| `TARGET_GROUP` | Demographic group targeted |
| `KEY_SPAN` | Exact text span that drove the decision |
| `EXPLANATION` | One-sentence reasoning |

The prompt we used embeds the Knight et al. (2017) definitions directly, including the facilitation clause and the shared-features warning.

````python
def classify_extremism(text, model_name=WORKING_MODEL):
    """
    Classifies a social media post into multiple extremism-related dimensions:
    - VE vs NVE
    - VE subtype
    - target group
    - key evidence span
    - explanation

    Parameters:
    -----------
    text : str
        Input social media post.
    model_name : str
        Model identifier verified earlier for gateway compatibility.

    Returns:
    --------
    dict
        Structured classification output.
    """

    # Defensive programming:
    # Convert missing or non-string values into a safe string format.
    # This prevents failures when the dataset contains NaN entries.
    text = "" if pd.isna(text) else str(text)

    # Prompt design:
    # Definitions are grounded in Knight, Woodward & Lancaster (2017) to ensure
    # classifications are consistent with the empirical VE/NVE distinction.
    # The shared-features warning prevents the model from conflating ideological
    # language with violence, since both groups share grievances and outgroup hostility.
    prompt = f"""You are a content classifier trained in extremism research.
Analyze the following social media post and classify it across the following dimensions.

DEFINITIONS (based on Knight, Woodward & Lancaster, 2017):

VE (Violent Extremism): Content involving any act, or clear intent to commit
an act, constituting murder, attempted murder, assault, or serious physical
harm to persons or property in pursuit of an ideological goal. This includes
content that knowingly facilitates such violence by others, such as financing
attacks, supplying weapons, or operational planning. Content that funds or
logistically supports a violent act is VE even if the author did not personally
commit violence.

NVE (Non-Violent Extremism): Content that promotes, disseminates or supports
an extreme ideology without involving or facilitating direct physical harm.
This includes distributing extremist material, hate speech, fundraising for
extremist causes, or recruiting others to an extreme ideology, where there is
no direct link to planned or actual physical violence.

IMPORTANT: Both VE and NVE share ideological content, grievances and outgroup
hostility. These features alone do not distinguish the two. Classify based on
whether the content involves, plans or enables physical harm (VE), or whether
it is confined to ideological expression or material support without a direct
violence component (NVE).

1. VE_NVE:
   - VE (Violent Extremism)
   - NVE (Non-Violent Extremism)

2. VE_SUBTYPE (our extension of the Knight et al. framework for finer-grained
   classification of violent content):
   - Ideological
   - Political
   - Religious
   - N/A (if NVE)

3. TARGET_GROUP:
   Choose from:
   - Religious group, Ethnic/racial group, Political group, Gender,
     Nationality, Occupation
   - Or "Unclear"

4. KEY_SPAN:
   Copy exact words or phrases from the post that influenced your decision

5. EXPLANATION:
   One sentence explaining your reasoning, referencing whether physical
   harm, facilitation of violence, or ideological promotion drove the
   classification

Post: "{text}"

Return valid JSON in this format:
{{
  "VE_NVE": "...",
  "VE_SUBTYPE": "...",
  "TARGET_GROUP": "...",
  "KEY_SPAN": "...",
  "EXPLANATION": "..."
}}"""

    # API call:
    # The Responses API is used here because it is the most reliable
    # interface with the CMU AI Gateway configuration.
    response = client.responses.create(
        model=model_name,
        input=prompt,
        max_output_tokens=300
    )

    # Extract the text portion of the model response.
    # Some providers may return empty or differently structured outputs,
    # so this section is wrapped defensively.
    try:
        raw_text = response.output[0].content[0].text
        if raw_text is None:
            raise ValueError("Empty response")
        raw_text = raw_text.strip()
    except Exception:
        raise ValueError("Model returned no usable text")

    # Remove markdown code fences if present.
    raw_text = raw_text.replace("```json", "").replace("```", "").strip()

    # Handle explicit refusal cases without crashing the pipeline.
    # These cases are preserved for later analysis rather than discarded.
    if "cannot assist" in raw_text.lower():
        return {
            "VE_NVE": "UNKNOWN",
            "VE_SUBTYPE": "UNKNOWN",
            "TARGET_GROUP": "UNKNOWN",
            "KEY_SPAN": "N/A",
            "EXPLANATION": "Model refused to classify this input due to safety constraints"
        }

    # Parse the JSON output into a Python dictionary.
    try:
        parsed = json.loads(raw_text)
    except json.JSONDecodeError:
        raise ValueError(f"JSON parsing failed. Raw response:\n{raw_text}")

    # Normalize missing keys so that downstream dataframe creation remains stable.
    result = {
        "VE_NVE": parsed.get("VE_NVE", "UNKNOWN"),
        "VE_SUBTYPE": parsed.get("VE_SUBTYPE", "UNKNOWN"),
        "TARGET_GROUP": parsed.get("TARGET_GROUP", "UNKNOWN"),
        "KEY_SPAN": parsed.get("KEY_SPAN", "N/A"),
        "EXPLANATION": parsed.get("EXPLANATION", "No explanation provided")
    }

    return result
````

## Step 7 - Five-Example Pipeline Validation

Before running the full batch, test the function with a few examples to confirm the API call succeeds, the prompt produces well-formed JSON, and all expected output keys are present. It is important to check that the pipeline handles edge cases gracefully - including content filter blocks and model refusals - before committing to the full 150-row batch run.





````python
# Before running inference on the full balanced sample, this
# cell evaluates the pipeline on five examples. This helps
# confirm that API calls succeed, outputs are well-formed,and refusal/filter cases are handled gracefully.

preview_df = sample_df.head(5).copy()
preview_results = []

for idx, row in tqdm(preview_df.iterrows(), total=len(preview_df)):
    text = "" if pd.isna(row['Original_Message']) else str(row['Original_Message'])

    try:
        result = classify_extremism(text)

        # Attach original text and dataset label for inspection.
        result['Original_Message'] = text
        result['Extremism_Label'] = row['Extremism_Label']

        preview_results.append(result)

        # Print each result for qualitative checking before batch execution.
        print(f"\nExample {idx}:")
        print(result)

    except Exception as e:
        error_msg = str(e)

        # Distinguish content filter failures from other runtime failures.
        if "content_filter" in error_msg.lower():
            explanation = "Blocked by content moderation filter"
        else:
            explanation = f"{type(e).__name__}: {e}"

        print(f"Error at example {idx}: {explanation}")

        preview_results.append({
            'Original_Message': text,
            'Extremism_Label': row['Extremism_Label'],
            'VE_NVE': 'UNKNOWN',
            'VE_SUBTYPE': 'UNKNOWN',
            'TARGET_GROUP': 'UNKNOWN',
            'KEY_SPAN': 'N/A',
            'EXPLANATION': explanation
        })

preview_results_df = pd.DataFrame(preview_results)

print("\nPreview run shape:", preview_results_df.shape)
display(preview_results_df)
````

---

## Step 8 - Batch Classification

Run the classification function across all 150 examples in the balanced sample. Errors are caught per row so that a single failure does not abort the pipeline. Results are saved to `ve_nve_classifications.csv`.

````python
# Batch classification over the balanced sample
# This cell applies the classification function to each post in
# the balanced sample, stores the structured outputs, and saves
# the final results as a CSV for later analysis and submission.

results = []

for idx, row in tqdm(sample_df.iterrows(), total=len(sample_df)):
    # Convert to string in case of missing values
    text = "" if pd.isna(row['Original_Message']) else str(row['Original_Message'])

    try:
        # Run classification (already returns structured output)
        result = classify_extremism(text)

        # Add original message and label for later comparison
        result['Original_Message'] = text
        result['Extremism_Label'] = row['Extremism_Label']

        results.append(result)

    except Exception as e:
        error_msg = str(e)

        # Handle content filter vs other errors
        if "content_filter" in error_msg.lower():
            explanation = "Blocked by content moderation filter"
        else:
            explanation = f"{type(e).__name__}: {e}"

        print(f"Error at row {idx}: {explanation}")

        # Keep row so we don't lose alignment
        results.append({
            'Original_Message': text,
            'Extremism_Label': row['Extremism_Label'],
            'VE_NVE': 'UNKNOWN',
            'VE_SUBTYPE': 'UNKNOWN',
            'TARGET_GROUP': 'UNKNOWN',
            'KEY_SPAN': 'N/A',
            'EXPLANATION': explanation
        })

# Save results
results_df = pd.DataFrame(results)
results_df.to_csv('ve_nve_classifications.csv', index=False)

print("Saved! Shape:", results_df.shape)
display(results_df.head())
````

````python
import json
from openai import OpenAI
from google.colab import userdata

CMU_API_KEY = userdata.get("CMU_API_KEY")

class llama_model:
    def __init__(self, model_name="llama3-2-11b-instruct"):
        self.client = OpenAI(
            api_key=CMU_API_KEY,
            base_url="https://ai-gateway.andrew.cmu.edu/v1"
        )
        self.model_name = model_name

    def _generate_prompt(self, text, prompt_type):
        base_prompt = f"""Analyze the following social media post for its severity level related to extremism. Assign a severity level and provide a concise explanation for your decision.

Severity Levels:
- Low: The content is mildly concerning, possibly expressing strong opinions or disagreements, but does not promote harm, discrimination, or explicit extremist views.
- Medium: The content expresses hateful rhetoric, discrimination, or promotes extremist ideologies without explicit calls for violence. It might dehumanize groups or spread misinformation to foster animosity.
- High: The content directly incites violence, glorifies violent extremist acts, promotes terrorism, or contains explicit threats against individuals or groups.

Post: "{text}"

Return valid JSON in this format:
{{
  "SEVERITY_LEVEL": "...",
  "EXPLANATION": "..."
}}"""

        if prompt_type == "zero-shot":
            return base_prompt
        elif prompt_type == "cot":
            return f"First, let's break down the social media post and identify any cues related to extremist content, hate speech, or incitement. Consider the language used, the potential impact on target groups, and any direct or indirect calls to action. Based on this reasoning, assign a severity level (Low, Medium, or High) and provide an explanation.\n\n{base_prompt}"
        else:
            raise ValueError("Invalid prompt_type. Must be 'zero-shot' or 'cot'.")

    def classify_severity(self, text, prompt_type, temperature):
        text = "" if pd.isna(text) else str(text)
        prompt = self._generate_prompt(text, prompt_type)

        try:
            completion = self.client.responses.create(
                model=self.model_name,
                input=prompt,
                max_output_tokens=300,
                temperature=temperature
            )
            raw_text = completion.output[0].content[0].text.strip()

            # Remove markdown code fences if present.
            raw_text = raw_text.replace("```json", "").replace("```", "").strip()

            if "cannot assist" in raw_text.lower() or "refused" in raw_text.lower():
                return {
                    "SEVERITY_LEVEL": "UNKNOWN",
                    "EXPLANATION": "Model refused to classify due to safety constraints"
                }

            parsed = json.loads(raw_text)
            return {
                "SEVERITY_LEVEL": parsed.get("SEVERITY_LEVEL", "UNKNOWN"),
                "EXPLANATION": parsed.get("EXPLANATION", "No explanation provided")
            }
        except json.JSONDecodeError:
            return {
                "SEVERITY_LEVEL": "UNKNOWN",
                "EXPLANATION": f"JSON parsing failed: {raw_text}"
            }
        except Exception as e:
            return {
                "SEVERITY_LEVEL": "UNKNOWN",
                "EXPLANATION": f"API call failed: {e}"
            }
````

````python
from openai import OpenAI

# Define the model to use for clarity assessment (using gpt-4.1-mini as it's a verified working model)
LLM_ASSESSMENT_MODEL = "gpt-4.1-mini"

def assess_llm_clarity(text, target_audience="General public (8th grader)", model_name=LLM_ASSESSMENT_MODEL):
    """
    Uses an LLM to assess the clarity and readability of a given text for a target audience,
    returning a rating on a 1-5 scale and an explanation.
    """
    # Ensure the text is a string
    text = str(text)

    # Skip assessment for blocked narratives
    if "BLOCKED:" in text:
        return {"LLM_Clarity_Score": None, "LLM_Clarity_Explanation": "Not assessed (blocked narrative)"}

    prompt = f"""Evaluate the clarity and readability of the following text for a target audience of '{target_audience}'.
Rate it on a scale of 1 to 5 using the following rubric:
1 = Very unclear, confusing, jargon-heavy
2 = Somewhat unclear, hard to follow
3 = Moderately clear, understandable with effort
4 = Clear, easy to understand
5 = Very clear, immediately understandable

Text to evaluate: {text}

Provide the score and a brief explanation in JSON format:
{{
  "LLM_Clarity_Score": [1-5],
  "LLM_Clarity_Explanation": "..."
}}"""

    client = OpenAI(
        api_key=CMU_API_KEY,
        base_url="https://ai-gateway.andrew.cmu.edu/v1"
    )

    try:
        completion = client.chat.completions.create(
            model=model_name,
            messages=[{"role": "user", "content": prompt}],
            max_tokens=200,
            temperature=0.0 # Keep temperature low for deterministic assessment
        )
        raw_text = completion.choices[0].message.content.strip()
        raw_text = raw_text.replace("```json", "").replace("```", "").strip()

        parsed = json.loads(raw_text)
        return {
            "LLM_Clarity_Score": parsed.get("LLM_Clarity_Score", None),
            "LLM_Clarity_Explanation": parsed.get("LLM_Clarity_Explanation", "No explanation provided")
        }
    except Exception as e:
        error_msg = str(e)
        print(f"Error assessing clarity with LLM for text: {text[:50]}... Error: {error_msg}")
        return {"LLM_Clarity_Score": None, "LLM_Clarity_Explanation": f"LLM assessment failed: {error_msg}"}

# Apply LLM clarity assessment to the generated narratives
print("Applying LLM clarity assessment...")
# Filter out narratives that were blocked during generation before assessment, as they don't have actual content.
assessable_narratives_df = narratives_df[~narratives_df['counter_narrative'].str.contains("BLOCKED:", na=False)].copy()

llm_clarity_results = []
for idx, row in tqdm(assessable_narratives_df.iterrows(), total=len(assessable_narratives_df), desc="LLM Clarity Assessment"):
    result = assess_llm_clarity(row['counter_narrative'])
    llm_clarity_results.append(result)

llm_clarity_df = pd.DataFrame(llm_clarity_results)

# Add the LLM clarity results back to the main narratives_df
# Use 'index' to merge back to the original rows, ensuring alignment
narratives_df = narratives_df.merge(assessable_narratives_df.reset_index()[['index', 'original_post', 'persona_key', 'model_id']].merge(llm_clarity_df.reset_index(drop=True), left_index=True, right_index=True), on=['original_post', 'persona_key', 'model_id'], how='left')

print("LLM clarity assessment complete.")
display(narratives_df.head())
````
